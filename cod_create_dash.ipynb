```python
import zipfile
import os

def compress_directory_to_zip(zip_filename, directory):
    with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(directory):
            for file in files:
                file_path = os.path.join(root, file)
                zipf.write(file_path, os.path.relpath(file_path, os.path.join(directory, '..')))

# Пример использования
compress_directory_to_zip('C:/Users/idukh/OneDrive/Рабочий стол/ДИСК D/INNOSTAGE/sql_request/dashboard_export/dashboard_export.zip', 'C:/Users/idukh/OneDrive/Рабочий стол/ДИСК D/INNOSTAGE/sql_request/dashboard_export/dashboard_export')  
```
