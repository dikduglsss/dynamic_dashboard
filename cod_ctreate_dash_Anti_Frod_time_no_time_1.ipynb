```python
numeric_columns = df.select_dtypes(include='number')

correlation_matrix = numeric_columns.corr()
final_query = []
threshold = 0.5

for col1 in correlation_matrix.columns:
    for col2 in correlation_matrix.columns:
        if col1 != col2 and correlation_matrix[col1][col2] > threshold:
            query = (
                f"  SELECT \r\n"
                f"    '{col1}\\{col2}' AS name,\r\n"
                f"    AVG({col1}) AS average_value_1,\r\n"
                f"    MIN({col1}) AS min_value_1,\r\n"
                f"    MAX({col1}) AS max_value_1,\r\n"
                f"    AVG({col2}) AS average_value_2,\r\n"
                f"    MIN({col2}) AS min_value_2,\r\n"
                f"    MAX({col2}) AS max_value_2\r\n"
                f"  FROM '{table_name}'\r\n"
                f"  UNION\r\n"
            )
            final_query.append(query)

# Объединение всех запросов с последним SELECT без UNION
final_query_time_6 = ''.join(final_query).rstrip('UNION\r\n') + ';\r\n'
print(final_query_time_6)
```
